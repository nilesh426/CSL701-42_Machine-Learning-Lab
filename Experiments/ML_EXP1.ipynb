Task 1: Import NumPy, Pandas, Matplotlib.pyplot and Seaborn.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from google.colab import files
uploaded = files.upload()
df = pd.read_csv("Titanic-Dataset.csv")

print("Dataset Loaded Succcessfully")
```

Task 5: Check dataset dimensions using df.shape (Expected: 891 x 12)

Task 6: Display dataset information using df.info().

```python
print("\nDataset Information:")
print(df.info())
```

Task 7: Generate summary statistics using df.describe(); record mean, minimum and
maximum values of Age and Fare.


```python
print("\nSummary Statistics:")
print(df.describe())

# Record mean, minimum and maximum values of Age and Fare
print("\nAge Statistics")
print("Mean Age :", df["Age"].mean())
print("Minimum Age :", df["Age"].min())
print("Maximum Age :", df["Age"].max())

print("\nFare Statistics")
print("Mean Fare :", df["Fare"].mean())
print("Minimum Fare :", df["Fare"].min())
print("Maximum Fare :", df["Fare"].max())
```

Task 8: Identify missing values using df.isnull().sum()

```python
print("\nMissing Values:")
print(df.isnull().sum())
```

 Task 9: Fill missing Age values with the median.

```python
df["Age"] = df["Age"].fillna(df["Age"].median())

print("\nMissing Age Values After Filling:")
print(df["Age"].isnull().sum())
```

 Task 10: Filter surviving passengers (Survived == 1).

```python
survivors = df[df["Survived"] == 1]

print("\nSurviving Passengers:")
print(survivors.head())
```

Task 11: Convert Fare column to a NumPy array.

```python
fare_array = df["Fare"].to_numpy()

print("\nFare NumPy Array:")
print(fare_array)
```

Task 12: Calculate maximum and average Fare using np.max() and np.mean().

```python
max_fare = np.max(fare_array)
avg_fare = np.mean(fare_array)

print("\nMaximum Fare:", max_fare)
print("Average Fare:", avg_fare)
```

 Task 13: Plot histogram of Age.

```python
plt.figure(figsize=(8,5))
plt.hist(df["Age"], bins=20, color="skyblue", edgecolor="black")
plt.title("Age Distribution of Passengers")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

Task 14: Plot bar chart of Survival Count

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Survived", data=df, palette="Set2")
plt.title("Survival Count")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Number of Passengers")
plt.show()
```

Task 15: Plot scatter graph of Age vs Fare

```python
plt.figure(figsize=(8,5))
plt.scatter(df["Age"], df["Fare"], color="red", alpha=0.6)
plt.title("Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.grid(True)
plt.show()
```
